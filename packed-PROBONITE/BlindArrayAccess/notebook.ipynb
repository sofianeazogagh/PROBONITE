```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data10 = pd.read_csv('perf/perf(10,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data10["clients"] = data10.index + 1
data10_para = pd.read_csv('perf/perf2(10,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data10_para["clients"] = data10_para.index + 1

data20 = pd.read_csv('perf/perf(20,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data20["clients"] = data20.index + 1
data20_para = pd.read_csv('perf/perf2(20,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data20_para["clients"] = data20_para.index + 1

data30 = pd.read_csv('perf/perf(30,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data30["clients"] = data30.index + 1
data30_para = pd.read_csv('perf/perf2(30,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data30_para["clients"] = data30_para.index + 1

data40 = pd.read_csv('perf/perf(40,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data40["clients"] = data40.index + 1
data40_para = pd.read_csv('perf/perf2(40,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data40_para["clients"] = data40_para.index + 1

data50 = pd.read_csv('perf/perf(50,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data50["clients"] = data50.index + 1
data50_para = pd.read_csv('perf/perf2(50,1to50).txt',sep='=', header=None, usecols=[1], names=["time"])
data50_para["clients"] = data50_para.index + 1

# fig, ax = plt.subplots(1,2,figsize=(10,5))
# ax[0].plot(df1.clients, df1.time)
# ax[0].set_title("10 features max without omp")
# ax[0].legend()

# ax[1].plot(df2.clients, df2.time)
# ax[1].set_title("10 features max with omp")
# ax[1].legend()
# plt.show()

# plt.plot(df1.clients,df1.time,'g-',label='10 features max')


# df
```

```python
x = np.linspace(1,50,50)
y_ohs_20 = 1224*x
y_ohs_50 = 1249*x
y_ohs_10 = 1159*x
y_ohs_30 = 1230*x
y_ohs_40 = 1165*x
```

```python
plt.xlabel("Number of clients")
plt.ylabel("Time in ms")

plt.plot(data50.clients,data50.time,'b-',label='50 features max without omp')
plt.plot(data50_para.clients,data50_para.time,':b',label='50 features max with omp')
plt.plot(x, y_ohs_50, '--b', label = 'ground truth')

plt.legend()
plt.show()
```
